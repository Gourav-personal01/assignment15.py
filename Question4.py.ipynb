# Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
#Company Name: ABC Corporation
#Location: India
#Contact Detail: 999-999-9999

from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def company_page():
    return "Welcome to ABC Corporation"

@app.route('/')
def company_details():
    return "Company Name: ABC Corporation\nLocation: India\nContact Detail: 999-999-9999"

if __name__ == "__main__":
    app.run()
